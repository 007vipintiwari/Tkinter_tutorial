```python
import numpy as np
import pandas as pd 
```

```python
data= pd.read_csv('decisontree.csv')
data.columns
```

```python
data.head()
```

```python
data.rename(columns={'class:buys_computer': 'buys_computer'}, inplace=True)
```

```python
data.head()
```

```python
data.buys_computer=[1 if each == "yes" else 0 for each in data.buys_computer]
```

```python
data.head()
```

```python
y=data.buys_computer.values
```

```python
x_data=data.drop(["buys_computer"], axis=1)
```

```python
x_data.head()
```

```python
x_data = data[['age','income','student','credit_rating']]
```

```python
x_data
```

```python
x = x_data
x=pd.get_dummies(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=42)
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("score :",dt.score(x_test,y_test))
```
